```python
# 1. Importar librerÃ­as necesarias
import pandas as pd
import numpy as np

# 2. Cargar el dataset
df = pd.read_csv("data/survey_results_public.csv")


# 3. Verificar dimensiones y primeras filas
print("Shape:", df.shape)
df.head()
```

```python
# Mostrar nÃºmero de valores nulos por columna
valores_nulos = df.isnull().sum()
porcentaje_nulos = (valores_nulos / len(df)) * 100

# Mostrar en una tabla ordenada
valores_faltantes = pd.DataFrame({
    'Valores Nulos': valores_nulos,
    'Porcentaje (%)': porcentaje_nulos.round(2)
}).sort_values(by='Valores Nulos', ascending=False)

# Mostrar solo columnas con mÃ¡s del 30% de valores nulos
valores_faltantes[valores_faltantes['Porcentaje (%)'] > 30]
```

```python
# Eliminar columnas con mÃ¡s del 30% de nulos
columnas_a_eliminar = valores_faltantes[valores_faltantes['Porcentaje (%)'] > 30].index
df_clean = df.drop(columns=columnas_a_eliminar)

# Verificar nuevas dimensiones
print("Nueva forma del DataFrame:", df_clean.shape)
df_clean.head()
```

```python
# Crear nuevo DataFrame solo con las columnas clave desde el original
columnas_modelo = [
    "Country",
    "EdLevel",
    "YearsCodePro",
    "Employment",
    "RemoteWork",
    "ConvertedCompYearly"  # ðŸŽ¯ Variable objetivo: salario anual
]

df_modelo = df[columnas_modelo].copy()

# Ver primeras filas
df_modelo.head()
```

```python
# 1. Eliminar filas sin salario
df_modelo = df_modelo[df_modelo["ConvertedCompYearly"].notnull()]

# 2. Eliminar filas sin aÃ±os de experiencia
df_modelo = df_modelo[df_modelo["YearsCodePro"].notnull()]

# 3. Reiniciar Ã­ndices
df_modelo.reset_index(drop=True, inplace=True)

# 4. Mostrar forma y primeras filas
print("Shape despuÃ©s de limpiar:", df_modelo.shape)
df_modelo.head()
```

```python
def limpiar_experiencia(x):
    if x == "Less than 1 year":
        return 0.5
    elif x == "More than 50 years":
        return 51
    try:
        return float(x)
    except:
        return np.nan

# Aplicar limpieza
df_modelo["YearsCodePro"] = df_modelo["YearsCodePro"].apply(limpiar_experiencia)

# Eliminar filas que quedaron como NaN despuÃ©s de la conversiÃ³n
df_modelo = df_modelo[df_modelo["YearsCodePro"].notnull()]
df_modelo.reset_index(drop=True, inplace=True)

# Verificar
df_modelo["YearsCodePro"].describe()
```

```python
# 1. Simplificar nivel educativo
def limpiar_nivel(x):
    if "Bachelor" in x:
        return "Bachelor"
    elif "Master" in x:
        return "Master"
    elif "Professional" in x or "doctoral" in x or "Ph.D" in x:
        return "Doctorate"
    elif "Some college" in x or "Secondary" in x or "Primary" in x:
        return "NoDegree"
    else:
        return "Other"

df_modelo["EdLevel"] = df_modelo["EdLevel"].apply(limpiar_nivel)

# 2. Simplificar tipo de empleo (primera etiqueta si hay varias)
df_modelo["Employment"] = df_modelo["Employment"].apply(lambda x: x.split(";")[0].strip())

# 3. Rellenar nulos en RemoteWork con "Unknown" y convertir a categorÃ­as
df_modelo["RemoteWork"] = df_modelo["RemoteWork"].fillna("Unknown")

# 4. Codificar categÃ³ricas con OneHotEncoding
df_final = pd.get_dummies(df_modelo, columns=["Country", "EdLevel", "Employment", "RemoteWork"], drop_first=True)

# Verificar
print("Shape final:", df_final.shape)
df_final.head()
```

```python
# Guardar el DataFrame final con features limpias y codificadas
df_final.to_csv("data/datos_limpios_modelo.csv", index=False)
```
